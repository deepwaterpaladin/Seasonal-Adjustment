```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from openpyxl import Workbook
from openpyxl import load_workbook
wb = load_workbook(filename='prices.xlsx')
ws = wb.active

def get_total_energy_data(ws):
    date_col = ws["B"]
    total_energy_col = ws["C"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in total_energy_col[11:279]]})
    return df

def get_energy_data(ws):
    date_col = ws["B"]
    energy_col = ws["E"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in energy_col[11:279]]})
    return df

def get_metals_data(ws):
    date_col = ws["B"]
    metals_col = ws["F"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in metals_col[11:279]]})
    return df

def get_agricultural_data(ws):
    date_col = ws["B"]
    agricultural_col = ws["G"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in agricultural_col[11:279]]})
    return df

def get_fish_data(ws):
    date_col = ws["B"]
    fish_col = ws["H"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in fish_col[11:279]]})
    return df

def get_forestry_data(ws):
    date_col = ws["B"]
    forestry_col = ws["I"]
    df = pd.DataFrame({'Date': [i.value for i in date_col[11:279]], 'Price': [i.value for i in forestry_col[11:279]]})
    return df

def plot_energy_data(data):
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(data['Date'], data['Price'])
    #ax.set_title(f'{ws[]}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    return fig

def test_plot():
    func_lst = [get_total_energy_data(ws), get_energy_data(ws), get_metals_data(ws), get_agricultural_data(ws), get_fish_data(ws), get_forestry_data(ws)]
    for func in func_lst:
        plot_energy_data(func)

test_plot()


```
